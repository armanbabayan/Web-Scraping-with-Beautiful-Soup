```python
# Function for converting range into a number.

def conv_rang_to_num(x):
    if pd.isnull(x):
        return x
    else:
        parts = x.split(' ')
        if len(parts) == 3:
            return (float(parts[0]) + float(parts[2]))/2
        try:
            return float(x)           
        except:
            return x
```

```python
df['Time to maturity of technology (in years)'] = df['Time to maturity of technology (in years)'].apply(conv_rang_to_num)
```
